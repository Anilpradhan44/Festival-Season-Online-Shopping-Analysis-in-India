```python
import seaborn as sns
import matplotlib.pyplot as plt

fx = sns.countplot(
    x='Gender', hue='Gender', data=main_data,
    palette= 'Set1',
    legend=False
)

for bars in fx.containers:
    fx.bar_label(bars)
plt.show()
```

```python
#plotting a bar chart for Gender and it's count

sales_gen = sns.countplot(x = 'Gender', data = main_data)

for bars in sales_gen.containers:
    sales_gen.bar_label(bars)
```

```python
# Plotting a bar chart for gender vs total amount

sales_gen = main_data.groupby('Gender', as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
print(sales_gen)

sns.set(rc={'figure.figsize': (15, 5)})
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen, palette = 'Set1', hue = 'Gender')
plt.show()
```

```python
*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*
```

```python
# plotting a bar chart for gender vs total amount
sales_gen = main_data.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

```python
*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*
```
